```python
class ListItem:
    def __init__(self, value):
        self.value = value
        self.next = None
        
class Queue:
    def __init__(self):
        self.first = None
        self.last = None
    
    def is_empty(self):
        return self.first == None

    def leave(self):
        if self.is_empty():
            return None
        l = self.last
        if l.next == None:
            self.last = None
            self.first = None
            return l.value
        else:
            while l.next.next != None:
                    l = l.next
            o = l.next
            l.next = None
            self.first =l
            return o.value
    
    def print_queue(self):
        p = self.last
        while p is not None:
            print(p.value, end=', ')
            p = p.next
        print()
    
    def enter(self, v):
        new_node = ListItem(v)
        current = self.first
        if self.first == None:
            self.first = new_node
        else:
            while current.next != None:
                current = current.next
            current.next = new_node
    
    def replace(self, i, j):
        value_i = None
        value_j = None
        current = self.first
        counter_i = 0
        counter_j = 0
        while value_i != i and value_j != j:
            if current.value == i:
                value_i = current.value
            else:
                counter_i += 1
                current = current.next
            if current.value == j:
                value_j = current.value
            else:
                counter_j += 1
                current = current.next
        
        current = self.first
        for i in range(counter_i):
            current = current.next
        self.value = value_j
            
        for j in range(counter_j):
            current = current.next
        self.value = value_i
```
